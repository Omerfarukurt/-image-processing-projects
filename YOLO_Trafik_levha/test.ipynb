```python
from ultralytics import YOLO
import cv2
```

```python
model=YOLO(r"C:\dl_proje\image_process\YOLO_Trafik_levha\runs\detect\traffic_sign_model\weights\best.pt")
```

```python
#test edilecek görselin yüklenmesi
image_path=r"C:\dl_proje\image_process\YOLO_Trafik_levha\traffic-sign-detection-yolov8-11\test\images\23f60eceb79dfc215da2b5af312b2382_jpg.rf.e811bac3a2dc13ad25fd39b0e531c649.jpg"
image=cv2.imread(image_path)
```

```python
#image tahmini
results=model(image_path)[0]
print(results)
```

```python
#kutu çizimi
for box in results.boxes:

    #kordinatlar
    x1,y1,x2,y2 =map(int,box.xyxy[0])
    cls_id=int(box.cls[0]) #classification id
    confidence=float(box.conf[0])
    label=f"{model.names[cls_id]} conf: {confidence:.2f}"

    
    cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),2)
    #etiketi image üzerine ekle
    cv2.putText(image,label,(x1,y1-10),cv2.FONT_HERSHEY_COMPLEX,0.6,(0,255,0),2)
cv2.imshow("prediction",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
